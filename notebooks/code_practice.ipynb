```python
#数据的预处理操作，预处理一般包括数据增强，尺寸统一等等
#预处理的必要操作：TtTensor(),Normalize()
class DataTransformer:
    #进行数据预处理
    def __init__(self,size=28):
        self.transform=transforms(
            transforms.ToTensor(),#通过这个函数，将数据转化成神经网络需要的数据格式
            transforms.Normalize(mean=[0.1307], std=[0.3081])#进行数据放缩，转化到标准的正态分布格式，有利于防止梯度爆炸
        )
    def __call__(self,x):
        return self.transform(x)
```

```python
#模型的骨架构建，一般是通过nn.Sequential
class MiniMlp(nn.Module):
    def __init__(self,input_num,output_num):
        super.__init__()
        self.mlp=nn.Sequential(
            nn.Flatten(),#展平除了通道以外的所有维度
            nn.Linear(784,128),
            nn.ReLU(),
            nn.Linear(128,64),
            nn.ReLU(),
            nn.Linear(64,10)
        )
    def forward(self,x):
        return self.mlp(x) #定义前向传播：直接让数据x通过mlp网络
```
